# code 总结
- **Function**

> `get_cookies`:
> - cookies更新，每次爬取之前，从网站copy新的cookies到.txt文件
> - Cookies_web.txt 和 Cookies_mobile.txt
> - get_cookies：读取.txt——>requests.get()
> - 全局对象：cookies_web cookies_mobile

> `initialize_spider`
> - 读取spider_history.txt，返回当前history的url_list,page_list[:-1],time_list[:-1]

> `record_spider_history`
> - 将爬取到的信息写入spider_history文件

> `find_weibo_id`
> - 依据从不实信息列表中的获得url_list,get到每一条不实信息的界面html
> - 再从该html中正则到"原文"的url——mobile版的个人微博界面
> - 根据该url获得个人微博界面的html——一系列信息，比如weibo id等

> `get_rumor_web_url`
> - 传入不实信息列表page，返回该page的list信息
> - 其中嵌入find_weibo_id函数，故返回了所有我们需要的信息

> `main`部分（并未定义main函数）
> - 定义爬取微博数目interval 定义写入的filename
> - 首先初始化，读取history，如果文件为空，需要自定义return的参数——try-except
> - 定义爬取微博flag变量——i
> - 进入while true循环
> - 依据history的page，get到对应网页的所有info
> - 进入判断
> - 判断：i与interval的比较 循环结束——break语句
> - 判断：当前page的最后一行晚于history，说明history已经有了此页全部信息，继续爬取下一行，page+1-------问题？如果此页面更新了。。
> - 否则，遍历当前爬取page的url，是否在history，如果在，继续遍历，直至遍历到新的url或者page+1，从新的url开始截取list信息，使用record_spider_history函数将截取的信息放入history文件中，此时更新i（更新history文件时便是更新i时）

- **code framework**

> `pipeline`
> - initialize ——> get page info ——> record page info ——> break
> - initialize: `initialize_spider`
> - get page info: `get_cookies`  `find_weibo_id`  `get_rumor_web_url`
> - record page info: `record_history`
> - break: `main`
